```python
#importação/tratamento
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea


#link do dataset utilizado: https://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+(diagnostic)
#importação de dados do dataset:
breastC = pd.read_csv("data.csv", encoding='utf-8')

```

```python
#exibindo dataset
breastC
```

```python
#contando valores dentro de "diagnosis" dividido por B-Benign & M-malign
breastC.diagnosis.value_counts()
```

```python
#B -> benign /// separando dataset por dados de tumores benigns
B = breastC.query("diagnosis == 'B'")
B.head()
```

```python
B = breastC.query("diagnosis == 'B'")
B.head()
```

```python
#M -> maligns /// separando dataset por dados de tumores malignos
M = breastC.query("diagnosis == 'M'")
M.head()
```

```python
#média 
meanRadiusM = M.fractal_dimension_mean.mean()
meanRadiusB = B.fractal_dimension_mean.mean()
#mediana 
medianRadiusM = M.fractal_dimension_mean.median()
medianRadiusB = B.fractal_dimension_mean.median()
#variância
varRadiusM = M.fractal_dimension_mean.var()
varRadiusB = B.fractal_dimension_mean.var()
#desvio-padrão
stdRadiusM = M.fractal_dimension_mean.std()
stdRadiusB = B.fractal_dimension_mean.std()

print(">*******************************************************<")
print(f"Média do fractal de um tumor maligno: {meanRadiusM:.6f}")
print(">-------------------------------------------------------<")
print(f"Mediana do fractal de um tumor maligno: {medianRadiusM:.6f}")
print(">-------------------------------------------------------<")
print(f"Desvio-padrão do fractal de um tumor maligno: {stdRadiusM:.6f}")
print(">-------------------------------------------------------<")
print(f"Variância do fractal de um tumor maligno: {stdRadiusM:.6f}")
print(">-------------------------------------------------------<")
print(f"Valor mínimo do fractal de tumor maligno: {M.fractal_dimension_mean.min()}")
print(">-------------------------------------------------------<")
print(f"Valor máximo do fractal de tumor maligno: {M.fractal_dimension_mean.max()}")
print(">-------------------------------------------------------<")
print(f"Média do fractal de um tumor benigno: {meanRadiusB:.6f}")
print(">-------------------------------------------------------<")
print(f"Mediana do fractal de um tumor benigno: {medianRadiusB:.6f}")
print(">-------------------------------------------------------<")
print(f"Desvio-padrão do fractal de um tumor benigno: {stdRadiusB:.6f}")
print(">-------------------------------------------------------<")
print(f"Variância do fractal de um tumor benigno: {stdRadiusB:.6f}")
print(">-------------------------------------------------------<")
print(f"Valor mínimo do fractal de tumor benigno: {B.fractal_dimension_mean.min()}")
print(">-------------------------------------------------------<")
print(f"Valor máximo do fractal de tumor benigno: {B.fractal_dimension_mean.max()}")
print(">-------------------------------------------------------<")
print(f"fractal máximo (geral): {breastC.fractal_dimension_mean.max()}")
print(">-------------------------------------------------------<")
print(f"fractal mínimo (geral): {breastC.fractal_dimension_mean.min()}")
print(">*******************************************************<")
```

```python
#agrupamento de dados
malignDiagnosis = M.diagnosis.shape[0]
benignDiagnosis = B.diagnosis.shape[0]
```

```python

plt.figure(1, figsize=(5, 5))
plt.bar(x=["Benignos", "Malignos"], height=[benignDiagnosis, malignDiagnosis], width=0.75, edgecolor='black')
plt.ylabel('Frequencia Absoluta', fontsize = 11)
plt.title("Contagem de tumores B e M", fontsize=11)
plt.show()
```

```python
#definindo faixa de valores de raio de tumor para cálculo da média
rangeBreastC = breastC.query("fractal_dimension_mean >= 0 and fractal_dimension_mean <= 1") ##def. range
radius = rangeBreastC.fractal_dimension_mean
radius.head()
```

```python
plt.title("Média no fractal de tumores")
plt.xlabel("valor médio")
plt.ylabel("Frequência absoluta")
plt.hist(radius, rwidth=0.9, edgecolor='black')
```

```python
sea.boxplot(x=radius)
```

```python
sea.displot(x=radius, kde=True, rug=True)
```
